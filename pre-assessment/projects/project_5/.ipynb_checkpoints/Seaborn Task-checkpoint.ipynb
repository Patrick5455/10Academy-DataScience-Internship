```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
task_df = pd.read_csv("pandas_tasks.csv")
task_df.head()
```

#### Question 1. Make a scatter plot of the Longitude and Latitude columns

```python
location_df = task_df[['Latitude', 'Longitude']]
location_df.head()
```

```python
x=location_df.Latitude.values.tolist()
y = location_df.Longitude.values.tolist()
```

```python
plt.subplots(figsize=(10,8))
plt.title('Scatter Plot of Latitiude and Longitude')
sns.scatterplot(x, y, alpha=.5, color='g');
```
